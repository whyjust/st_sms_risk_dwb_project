```python
rule_map = {
    'A01':'(?:消费|下单).*?元|申请办理贷款|开通成功|成功发放|已获得.*?额度|成功开通|申请办理贷款',
    'A02':'未通过|未放款成功|放款失败',
    'A03':'已[结还]清|成功还款|还款成功|满足最低还款额',
    'A04':'本期应还|还款提醒|及时还款|还款卡余额充足|今日需还款',
    'A05':'(?:还款|扣款|扣划)失败',
    'A06':'已[逾超]期|未按时还款|已过还款日',
    'A07':'严重逾期|催缴|追缴|拒绝|严重违约|恶意违约|债权方|逃废债|诉前调查组|逾期材料|冻结全款',
    'A08':'最高.*?额度|免费额度|距离提现|领取[您你]的[0-9]+元|确认提款|提额福利|优惠利率|立即提现|额度.*?失效',
}

content = '【携程金融】消费1434.83元，请确保银行卡余额充足，将于12月23日优先从尾号5884银行卡自动扣款，详情请前往钱小乐APP查看，已还请忽略'
for k, v in rule_map.items():
    # print(f"when content regexp '{v}' then '{k}'")
    full_pattern = re.compile(v)
    full_result = full_pattern.findall(content)

    if len(full_result) > 0:
        print(f"{v} then {k}")
        print(full_result)
```
